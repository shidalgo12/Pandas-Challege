```python
# Dependencies and Setup
import pandas as pd

# File to Load
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df
```

```python
player_df = pd.DataFrame(purchase_df.drop_duplicates("SN"))
player_df
```

```python
# Player Count - COMPLETE

total_players = player_df.count()[0]

player_count_df = pd.DataFrame({'Total Players':[total_players]})
player_count_df
```

```python
# Purchasing Analysis (Total) - COMPLETE
    # Table: Number of Unique Items, Average Price, Number of Purchases & Total Revenue

# Number of Unique Items
total_items = len(purchase_df["Item ID"].unique())
# total_items

# # Average Price
avg_price = round(purchase_df["Price"].mean(),2)
# avg_price

# Number of Purchases
total_purchases = len(purchase_df["Purchase ID"].unique())
# total_purchases

# Total Revenue
total_revenue = purchase_df["Price"].sum()
# total_revenue
    
# Create a summary data frame to hold the results
purchase_total_data = {
    'Number of Unique Items':[total_items],
    'Average Price': [avg_price],
    'Number of Purchases': [total_purchases],
    'Total Revenue': [total_revenue]}

purchase_totals_df = pd.DataFrame(purchase_total_data)


# Format Data
purchase_totals_df["Average Price"] = purchase_totals_df["Average Price"].map("${:,.2f}".format)
purchase_totals_df["Total Revenue"] = purchase_totals_df["Total Revenue"].map("${:,.2f}".format)

purchase_totals_df
```

```python
# Gender Demographics - COMPLETE
    # Table : Total Count , Percentage of Players

# Calculate Players per gender
gender_count = player_df["Gender"].value_counts()
# gender_count

# Calculate Percentage of player per gneder
gender_percentage = gender_count/total_players
# gender_percentage

# Display the summary data frame
gender_demographics = pd.DataFrame({"Total Count": gender_count, "Percentage of Players": gender_percentage})

# Format Data
gender_demographics['Percentage of Players'] = gender_demographics['Percentage of Players'].map("{:,.2%}".format)
gender_demographics
```

```python
# Gender Purchasing Analysis
    # Table: Purchase Count, Average Purchase Price, Total Purchase Value, Avg Total Purchase Per Person 

# Calculate total players per gender
gender_counts = purchase_df.groupby(["Gender"]).count()["Price"].rename("Purchase Count")
#Calculate average purchase value per gender
gender_average = round(purchase_df.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price"),2)

# Calculate total purchases per gender
gender_purchase_total = player_df.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

# Calculate Average Total Purchase per Person
avg_total_purchase_per_person = round(gender_purchase_total / gender_demographics["Total Count"],2)

# Convert to DataFrame
gender_purchases = pd.DataFrame({"Purchase Count": gender_counts, "Average Purchase Price": gender_average, "Total Purchase Value": gender_purchase_total, "Avg Total Purchase Per Person": avg_total_purchase_per_person})

# Format Data
gender_purchases ["Purchase Count"] = gender_purchases["Purchase Count"].map("{:,}".format)
gender_purchases["Average Purchase Price"] = gender_purchases["Average Purchase Price"].map("${:,.2f}".format)
gender_purchases["Total Purchase Value"] = gender_purchases["Total Purchase Value"].map("${:,.2f}".format)
gender_purchases["Avg Total Purchase per Person"] = gender_purchases["Avg Total Purchase Per Person"].map("${:,.2f}".format)
gender_purchases = gender_purchases.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
gender_purchases
```

```python
#Age Demographics - COMPLETE
    # Table: Total Count, Percentage of Players

# Establish bins for ages
bins = (0,9,14,19,24,29,34,39,45)
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Add "Age Group" Column to player_df
player_df["Age Group"] = pd.cut(player_df["Age"], bins, labels=group_names)
# player_df


# Number of Players Per Age Group
age_count = player_df["Age Group"].value_counts()
# age_count

# Percentage of Players Per Age Group
age_percentage = age_count/total_players
# age_percentage

# Build summary data table
age_demographics = pd.DataFrame({'Total Count': age_count,'Percentage of Players': age_percentage})
age_demographics['Percentage of Players'] = age_demographics['Percentage of Players'].map("{:,.2%}".format)
# age_demographics

# Rearrange order of rows under Age Group
age_demographics = pd.DataFrame(age_demographics, index=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"])
age_demographics = age_demographics.reset_index()
age_demographics = age_demographics.rename(columns= {'index':'Age Group'})


age_demographics
```

```python
# Top Spenders - COMPLETE
    # Table headers: Purchase Count, Average Purchase Price, Total Purchase Value

# Total purchases per player
total_player_purchases = purchase_df.groupby(['SN']).count()["Purchase ID"].rename("Purchase Count")
total_player_purchases

# Average purchase value per player
avg_player_spending = round(purchase_df.groupby(['SN']).mean()["Price"].rename("Average Purchase Price"),2)
avg_player_spending

# Total spending per player
total_player_spending = purchase_df.groupby(['SN']).sum()["Price"].rename('Total Purchase Value')
total_player_spending

# Convert to DataFrame
player_purchases = pd.DataFrame({"Purchase Count": total_player_purchases, "Average Purchase Price": avg_player_spending, "Total Purchase Value": total_player_spending})
player_purchases

# Sort Total Purchase Value in descending order
sorted_player_purchases = player_purchases.sort_values("Total Purchase Value", ascending=False)

# Filter by Top 5 Players
sorted_player_purchases = sorted_player_purchases.nlargest(5,["Total Purchase Value"])

# Format Data
sorted_player_purchases["Average Purchase Price"] = sorted_player_purchases["Average Purchase Price"].map("${:,.2f}".format)
sorted_player_purchases["Total Purchase Value"] = sorted_player_purchases["Total Purchase Value"].map("${:,.2f}".format)
sorted_player_purchases
```

```python
# Most Popular Items
    # Table: Item ID, Item Name, Purchase Count, Item Price, Total Purchase Value

# Retrieve the Item ID, Item Name, and Item Price columns
item_df = pd.DataFrame(purchase_df[["Item ID","Item Name","Price"]])
item_df
    
# Total purchases per item
item_purchase_count = purchase_df.groupby(["Item ID","Item Name"]).count()["Price"].rename("Purchase Count")
item_purchase_count

# Item price
item_price = purchase_df.groupby(["Item ID","Item Name"]).mean()["Price"].rename("Item Price")
item_price

# Total purchase value per item
total_purchase_value = purchase_df.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Purchase Value")
total_purchase_value

# Convert to DataFrame
popular_items = pd.DataFrame({"Purchase Count": item_purchase_count, "Item Price": item_price, "Total Purchase Value": total_purchase_value})
popular_items

# Sort purchase count in descending order
sorted_popular_items = popular_items.sort_values("Purchase Count", ascending=False)
# Filter by Top 5 Selling Items
sorted_popular_items = sorted_popular_items.nlargest(10,["Purchase Count"])

# Format Data
sorted_popular_items["Item Price"] = popular_items["Item Price"].map("${:,.2f}".format)
sorted_popular_items["Purchase Count"] = popular_items["Purchase Count"].map("{:,}".format)
sorted_popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map("${:,.2f}".format)

sorted_popular_items
```

```python
# Most Profitable Items - COMPLETE

# Sort the above table by total purchase value in descending order
profitable_items = popular_items.sort_values("Total Purchase Value", ascending=False)
profitable_items = profitable_items.nlargest(5,["Total Purchase Value"])

profitable_items["Item Price"] = profitable_items["Item Price"].map("${:,.2f}".format)
profitable_items["Total Purchase Value"] = profitable_items["Total Purchase Value"].map("${:,.2f}".format)

profitable_items
```
