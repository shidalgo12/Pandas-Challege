```python
# Dependencies and Setup
import pandas as pd

# File to Load
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
# Player Count - COMPLETE

total_players = len(purchase_data["SN"].unique())

player_count_data = {'Total Players':[total_players]}
player_count_df = pd.DataFrame(player_count_data)
player_count_df
```

```python
# Purchasing Analysis (Total) - COMPLETE
    # Table: Number of Unique Items, Average Price, Number of Purchases & Total Revenue

# Number of Unique Items - COMPLETE
total_items = len(purchase_data["Item ID"].unique())
# total_items

# Average Price - COMPLETE
avg_price = round(purchase_data["Price"].mean(),(2))
# avg_price

# Number of Purchases - COMPLETE
total_purchases = len(purchase_data["Purchase ID"].unique())
# total_purchases

# Total Revenue - COMPLETE
total_revenue = purchase_data["Price"].sum()
# total_revenue
    
# Create a summary data frame to hold the results
purchase_total_data = {
    'Number of Unique Items':[total_items],
    'Average Price': [avg_price],
    'Number of Purchases': [total_purchases],
    'Total Revenue': [total_revenue]}

purchase_df = pd.DataFrame(purchase_total_data)
purchase_df
# # Use Map to format all the columns
# file_df["avg_cost"] = file_df["avg_cost"].map("${:.2f}".format)
# file_df["population"] = file_df["population"].map("{:,}".format)
# file_df["other"] = file_df["other"].map("{:.2f}".format)
# file_df.head()

    # Optional: give the displayed data cleaner formatting
    
# Display the summary data frame
```
